```python
import csv

def generate_thrust_curve_csv(duration, thrust=3500, timestep=0.01, filename="thrust_curve.csv"):
    """
    Generates a CSV file representing a constant thrust curve.

    :param duration: total burn time (in seconds)
    :param thrust: constant thrust value (in Newtons)
    :param timestep: time step between data points (in seconds)
    :param filename: output CSV file name
    """
    # Calculate the total number of steps
    steps = int(duration / timestep) + 1

    # Open the CSV file for writing
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        # Write header row
        writer.writerow(["Time (s)", "Thrust (N)"])
        
        # Write thrust values at each time step
        for i in range(steps):
            time = round(i * timestep, 3)
            writer.writerow([time, thrust])
    
    print(f"✅ File '{filename}' generated with constant thrust of {thrust} N for {duration} seconds.")

# Ask user to enter the duration
try:
    user_duration = float(input("Enter thrust duration (in seconds): "))
    if user_duration <= 0:
        raise ValueError
    generate_thrust_curve_csv(user_duration)
except ValueError:
    print("⛔ Please enter a valid duration (> 0).")
```
