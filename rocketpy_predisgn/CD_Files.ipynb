```python
def convertir_csv_europeen_vers_us(input_path, output_path):
    valeurs_deja_vues = set()
    
    with open(input_path, "r", encoding="utf-8") as f_in:
        lignes = f_in.readlines()

    with open(output_path, "w", encoding="utf-8") as f_out:
        for i, ligne in enumerate(lignes):
            ligne_modifiee = ligne.replace(",", ".").replace(";", ",").strip()
            
            if not ligne_modifiee:
                continue

            premiere_valeur = ligne_modifiee.split(",")[0]

            if premiere_valeur in valeurs_deja_vues:
                break

            valeurs_deja_vues.add(premiere_valeur)
            f_out.write(ligne_modifiee + "\n")
            

# Exemple d'utilisation
convertir_csv_europeen_vers_us("CD_Power_Off.csv", "CD_Power_Off_us.csv")
convertir_csv_europeen_vers_us("CD_Power_On.csv", "CD_Power_On_us.csv")
```
